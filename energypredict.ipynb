```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler

```

```python
data = pd.read_csv("energy.csv")
data.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data):
    # Drop duplicate rows across all columns
    data = data.drop_duplicates()
    # Drop rows with missing data across all columns
    data = data.dropna()
    return data

data = clean_data(data)
data.head()
```

```python
indexx = data["X7"].index
indexx
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data[:,7].shape
```

```python
X = scaled_data[:,[0, 1, 2, 3, 4,5,6,7]]
X = X.astype("float")

X.ndim

type(X)
```

```python
scaled_data[:,8:10].shape
```

```python
y1 = scaled_data[:,[8]]
y1 = y1.astype("float")
type(y1)
y1.shape
```

```python
y2 = scaled_data[:,[9]]
y2.shape
y2
```

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
# Veriyi eğitim ve test verilerine bölün
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor


# split the dataset into training and testing sets
X_train, X_test, y1_train, y1_test = train_test_split(X, y1, test_size = 0.30, random_state=77)
X_train, X_test, y2_train, y2_test = train_test_split(X, y2, test_size = 0.30, random_state=77)


# train the model and make predictions for heating_load
lr_heating = LinearRegression(n_jobs=-1)
y1_pred = lr_heating.fit(X_train, y1_train).predict(X_test)

# train the model and make predictions for cooling_load
lr_cooling = LinearRegression(n_jobs=-1)
y2_pred = lr_cooling.fit(X_train, y2_train).predict(X_test)

# calculate the MSE
print('MSE of predicting heating load: \n', mean_squared_error(y1_test, y1_pred).round(4))
print('MSE of predicting cooling load: \n', mean_squared_error(y2_test, y2_pred).round(4))
print('')


# calculate the RMSE
print('RMSE of predicting heating load: \n', mean_squared_error(y1_test, y1_pred, squared=False).round(4))
print('RMSE of predicting cooling load: \n', mean_squared_error(y2_test, y2_pred, squared=False).round(4))
print('')
```

```python
# calculate the MAE
from sklearn.metrics import mean_absolute_error


print('MAE of predicting heating load: \n', mean_absolute_error(y1_test, y1_pred).round(2))
print('MAE of predicting cooling load: \n', mean_absolute_error(y2_test, y2_pred).round(2))
```
